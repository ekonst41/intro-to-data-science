При выполнении каждой итерации с целью дальнейшего анализа сохраняйте текущее значение оптимизируемого функционала, а также затраченное время на итерацию.  **При реализации класса запрещено пользоваться ИИ-инструментами.**

4. Следите за качеством кода, комментируйте логические этапы кода. Несоблюдение этого требования может привести к потере баллов.


```python
# При реализации класса запрещено пользоваться ИИ-инструментами.


class LogisticRegression(BaseEstimator):
    """Модель логистической регрессии.

    Параметры:
    method (Literal['gd', 'sgd']): Метод оптимизации ('gd' - градиентный спуск,
        'sgd' - стохастический градиентный спуск).
    learning_rate (float): Константа скорости обучения, на которую домножаем градиент при обучении
    tol (float): Допустимое изменение функционала между итерациями.
    max_iter (int): Максимальное число итераций.
    batch_size (int): Размер выборки для оценки градиента (используется только при 'sgd').
    fit_intercept (bool): Добавлять ли константу в признаки.
    save_history (bool): Сохранять ли историю обучения.
    """

    def __init__(
        self,
        method: Literal["gd", "sgd"] = "gd",
        learning_rate: float = 0.5,
        tol: float = 1e-3,
        max_iter: int = int(1e4),
        batch_size: int = 64,
        fit_intercept: bool = True,
        save_history: bool = True,
    ):
        """Создает модель и инициализирует параметры."""
        self.method = method
        self.learning_rate = learning_rate
        self.tol = tol
        self.max_iter = max_iter
        self.batch_size = batch_size
        self.fit_intercept = fit_intercept
        self.save_history = save_history
        self.history = []  # История обучения

    @staticmethod
    def _sigmoid(x: np.ndarray) -> np.ndarray:
        """Вычисляет сигмоидную функцию."""
        return 1 / (1 + np.exp(-x))

    def _add_intercept(self, X: np.ndarray) -> np.ndarray:
        """Добавляет свободный коэффициент к матрице признаков.

        Параметры: X (np.ndarray): Исходная матрица признаков.

        Возвращает: np.ndarray: Матрица X с добавленным свободным
        коэффициентом.
        """
        X_copy = np.full((X.shape[0], X.shape[1] + 1), fill_value=1)
        X_copy[:, :-1] = X
        return X_copy

    def loss_count(self, X: np.ndarray, Y: np.ndarray):
        """
        Логистическая функция потерь
        """
        z = X.dot(self.coef_)
        loss = np.sum(np.logaddexp(0, z) - Y * z) / X.shape[0]
        return loss

    def fit(self, X: np.ndarray, Y: np.ndarray) -> "LogisticRegression":
        """Обучает модель логистической регрессии.

        Также, в случае self.save_history=True, добавляет в self.history
        текущее значение оптимизируемого функционала и затраченное время.

        Параметры:
        X (np.ndarray): Матрица признаков.
        Y (np.ndarray): Вектор истинных меток.

        Возвращает:
        LogisticRegression: Обученная модель.
        """
        if X.shape[0] != Y.shape[0]:
            raise ValueError("Количество строк в X и Y должно совпадать")

        if self.fit_intercept:
            X_copy = self._add_intercept(X)
        else:
            X_copy = X.copy()

        # начальные параметры модели
        n_samples, n_features = X_copy.shape
        self.coef_ = np.zeros(n_features)
        prev_loss = np.inf
        start_time = time()
        iter_count = 0

        # итерация для градиентного спуска
        for iter_count in range(self.max_iter):
            if self.method == 'sgd':
                index = np.random.choice(n_samples, self.batch_size, replace=False)
                X_batch, Y_batch = X_copy[index], Y[index]
            else:
                X_batch, Y_batch = X_copy, Y
            x = X_batch.dot(self.coef_)
            Y_pred = self._sigmoid(x)
            grad = X_batch.T.dot(Y_pred - Y_batch) / Y_batch.size
            self.coef_ -= self.learning_rate * grad
            loss = self.loss_count(X_copy, Y)

            current_time = time()
            iter_time = current_time - start_time
            start_time = current_time

            if self.save_history:
                self.history.append({"iter": iter_count, "loss": loss, "time": iter_time})

            if np.abs(loss - prev_loss) <= self.tol:
                break
            else:
                prev_loss = loss

        if self.fit_intercept:
            self.intercept_ = self.coef_[-1] # Свободный коэффициент
        else:
            self.intercept_ = 0.0
        self.n_iter_ = iter_count  # Число итераций

        return self

    def predict(self, X: np.ndarray) -> np.ndarray:
        """Возвращает предсказанные классы.

        Параметры: X (np.ndarray): Матрица признаков.

        Возвращает: np.ndarray: Предсказанные классы.
        """
        if self.fit_intercept:
            X_copy = self._add_intercept(X)
        else:
            X_copy = X.copy()

        if X_copy.shape[1] != self.coef_.shape[0]:
            raise ValueError("Число признаков в X не соответствует числу коэффициентов модели")

        x = X_copy.dot(self.coef_)
        predictions = (self._sigmoid(x) >= 0.5).astype(int)
        return predictions

    def predict_proba(self, X: np.ndarray) -> np.ndarray:
        """Возвращает вероятности классов 0 и 1.

        Параметры: X (np.ndarray): Матрица признаков.

        Возвращает: np.ndarray: Матрица вероятностей классов (n_samples,
        2).
        """
        if self.fit_intercept:
            X_copy = self._add_intercept(X)
        else:
            X_copy = X.copy()

        if X_copy.shape[1] != self.coef_.shape[0]:
            raise ValueError("Число признаков в X не соответствует числу коэффициентов модели")

        x = X_copy.dot(self.coef_)
        prob = self._sigmoid(x)
        return np.vstack((1 - prob, prob)).T
```

**2.** Обучите две модели логистической регрессии с помощью методов
* простой градиентный спуск;
* стохастический градиентный спуск.

```python
def average(histories):
    avg_history = []
    max_iters = max(len(h) for h in histories)
    for i in range(max_iters):
        losses = []
        times = []
        for h in histories:
            if i < len(h):
                losses.append(h[i]['loss'])
                times.append(h[i]['time'])
        avg_loss = np.mean(losses) if losses else np.nan
        avg_time = np.mean(times) if times else np.nan
        avg_history.append({'iter': i, 'loss': avg_loss, 'time': avg_time})
    return avg_history
```

```python
max_iter = 1000
learning_rate = 0.2
tol = 1e-6
batch_size = 64

gd_hist = []
sgd_hist = []

model_gd = LogisticRegression(method='gd', learning_rate=learning_rate, max_iter=max_iter, tol=tol, save_history=True)
model_sgd = LogisticRegression(method='sgd', learning_rate=learning_rate, max_iter=max_iter, tol=tol, batch_size=batch_size, save_history=True)

for i in range(7):
  model_gd.fit(X_train_scaled, y_train.values)
  model_sgd.fit(X_train_scaled, y_train.values)
  gd_hist.append(model_gd.history)
  sgd_hist.append(model_sgd.history)
  model_gd.history = []
  model_sgd.history = []

gd_history = average(gd_hist)
sgd_history = average(sgd_hist)
```

### Задача 6.

В этой задаче вам предлагается реализовать регрессию Хьюбера, а также применить ее к данным с выбросами. Для начала реализуйте класс по шаблону снизу. Обратите внимание, что класс `HuberRegression` &mdash; наследник класса `BaseEstimator`, это с легкостью позволит использовать наш класс в различных пайплайнах библиотеки `sklearn`.

**1.** Задача оптимизации для регрессии Хьюбера выглядит следующим образом:
$$\sum_{i=1}^n R\left(Y_i - x_i^T\theta\right) \rightarrow \min_\theta,$$
где $R(x)$ &mdash; функция потерь Хьюбера, определяемая как
$$R(x) = \frac{x^2}{2} I\left\{|x| < c\right\} + c \left(|x| - \frac{c}{2}\right)I\left\{|x| > c\right\}.$$

Выпишите формулы для градиентного и стохастического градиентного спусков. В чем польза такой функции потерь?

Функцию потерь удобнее выразить следующим образом:

$R(x) = \begin{cases}
\frac{x^2}{2} & \text{если } |x| < c, \\
c \left( |x| - \frac{c}{2} \right) & \text{если } |x| \geq c.
\end{cases}$

Пусть $r_i = Y_i - x_i^T \theta$

Тогда градиент функции $\nabla L = \sum\limits_{i=1}^{n}\nabla R(r_i) = \begin{cases}
- r_i x_i & \text{если } |x| < c, \\
-c ~sign(r_i) x_i& \text{если } |x| \geq c.
\end{cases}$

Шаг градиентного спуска:

$\theta_{t+1} = \theta_t - \eta \sum\limits_{i=1}^{n} \nabla R$

Для стохастического градиентного спуска выбирается случайный батч (набор элементов) и записывается то же самое выражение.

**2.** Реализуем теперь класс. **При реализации класса запрещено пользоваться ИИ-инструментами.**

```python
# При реализации класса запрещено пользоваться ИИ-инструментами.


class HuberRegression(BaseEstimator):
    """Класс, реализующий линейную регрессию с функцией потерь Хьюбера."""

    def __init__(self, c: float = 1.0, fit_intercept: bool = True, max_iter: int = 1000, learning_rate: float = 0.1, tol: float = 1e-4) -> None:
        """Инициализирует модель.

        Параметры: c (float): Константа из функции потерь Хьюбера.
        fit_intercept (bool): Добавлять ли константный признак. max_iter
        (int): Максимальное число итераций оптимизации.
        """
        self.c = c
        self.fit_intercept = fit_intercept
        self.max_iter = max_iter
        self.n_iter = 0
        self.learning_rate = learning_rate
        self.coef = None
        self.intercept = None
        self.tol = tol

    def huber_loss(self, r: np.ndarray) -> float:
        """Вычисляет значение функции потерь Хьюбера."""
        loss = np.where(np.abs(r) < self.c, 0.5 * r**2, self.c * (np.abs(r) - 0.5 * self.c**2))
        return np.sum(loss) / len(r)

    def fit(self, X: np.ndarray, y: np.ndarray) -> "HuberRegression":
        """Обучает модель.

        Параметры:
        X (np.ndarray): Матрица признаков.
        y (np.ndarray): Вектор целевой переменной.

        Возвращает:
        HuberRegression: Обученная модель.
        """
        if X.shape[0] != y.shape[0]:
            raise ValueError("Количество строк в X и y должно совпадать")

        if self.fit_intercept:
            X_copy = np.full((X.shape[0], X.shape[1] + 1), fill_value=1)
            X_copy[:, :-1] = X
        else:
            X_copy = X

        self.coef = np.zeros(X_copy.shape[1])
        self.intercept = 0
        prev_loss = 0

        for i in range(self.max_iter):
            r = y - X_copy.dot(self.coef)
            r_derivative = np.where(np.abs(r) < self.c, r, self.c * np.sign(r))
            grad = -X_copy.T.dot(r_derivative)
            grad /= len(y)

            self.coef -= self.learning_rate * grad
            loss = self.huber_loss(r)

            if np.abs(loss - prev_loss) <= self.tol:
              break
            else:
              prev_loss = loss

        if self.fit_intercept:
          self.intercept = self.coef[-1]
        self.n_iter_ = i + 1  # Число итераций

        return self

    def predict(self, X: np.ndarray) -> np.ndarray:
        """Делает предсказание на новых данных.

        Параметры: X (np.ndarray): Матрица признаков.

        Возвращает: np.ndarray: Вектор предсказанных значений.
        """

        if self.fit_intercept:
            X_copy = np.full((X.shape[0], X.shape[1] + 1), fill_value=1)
            X_copy[:, :-1] = X
        else:
            X_copy = X

        if X_copy.shape[1] != self.coef.shape[0]:
            raise ValueError("Число признаков в X не соответствует числу коэффициентов модели")

        pred = X_copy.dot(self.coef)
        return pred
```

### Задача 7.


Рассмотрим модель одномерной регрессии $y(x) = \theta x$, где $x \in \mathbb{R}$ &mdash; одномерный признак, $y \in \mathbb{R}$ &mdash; целевой признак, $\theta \in \mathbb{R}$ &mdash; неизвестный параметр. Имеется выборка размера $n$, полученная по правилу
	$$Y_i = \theta x_i + \varepsilon_i,\ \ \ i=1,...,n,$$
где $\varepsilon_i$ &mdash; случайная ошибка измерений.

Предложите точный алгоритм поиска оценки параметра $\theta$ методом наименьших модулей, то есть $$\sum_{i=1}^n \left|Y_i - \theta x_i\right| \to \min_\theta,$$ работающий за время $O(n \log n)$. Приведите его описание и теоретическое обоснование. Реализация в коде не требуется.

**Решение:**


1. Сортировка точек:

Отсортируем точки $(x_i, Y_i)$ по значению $\frac{Y_i}{x_i}$. Это эквивалентно сортировке по углу $\alpha = arctg(\frac{Y_i}{x_i})$, так как арктангенс монотонен.

Сложность сортировки: O(nlogn)

2. Инициализация:

Начнем с $\theta = -∞$. В этом случае все точки лежат в области положительных модулей.

Вычислим начальное значение функции потерь:

$L(\theta) = \sum\limits_{i=1}^{n} (Y_i - x_i \theta)$

Сложность: O(n)

3. Перебор точек:

Будем последовательно перебирать точки в порядке возрастания $\frac{Y_i}{x_i}$

Для каждой точки:
- Обновим значение $\theta = \frac{Y_i}{x_i}$
- Пересчитаем значение функции потерь $L(θ)$, учитывая, что точка
$(x_i, Y_i)$ переходит из области положительных модулей в ноль, а затем в отрицательные модули, т.е. $Y_i - x_i \theta \rightarrow 0 \rightarrow x_i \theta - Y_i$. Для каждого изменения $\theta$ необходимо сделать ровно 2 пересчёта, для предыдущей точки и текущей.

Итоговая апроксимация обязательно проходит через одну из точек, иначе, если мы провели прямую между 2 точками, но смещением до ближайшей точки можно уменьшить функционал.

Всего таких переходов по точкам будет n, для каждого перехода с функционалом необходимо совершить константное число действий.

Сложность: O(n)

4. Итог

В итоге можем из всех найденных значений функционала можно найти минимальный, при этом общая сложность O(nlogn)
